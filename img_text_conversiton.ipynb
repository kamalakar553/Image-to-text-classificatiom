image to text convertion 

```python
import pytesseract
import cv2
from IPython.display import display
from PIL import Image
import numpy as np

path="Image\istockphoto-1356112888-612x612.jpg"
```

```python

# Load the image

img = cv2.imread(path)
display(img.open(path))
image = np.array(img)
```

```python
# configurations
config = ('-l eng --oem 1 --psm 3')
```

```python
pytesseract.pytesseract.tesseract_cmd = r'Downloads/tesseract/tesseract.exe'
# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Apply OCR to extract text
text = pytesseract.image_to_string(gray,config=config)
```

```python
import joblib
loaded_model = joblib.load(open('model.pkl', 'rb'))
x=loaded_model.predict([text])
x[0]
```
